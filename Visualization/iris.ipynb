```python
from sklearn import datasets
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
# Load the IRIS dataset from standard libraries

iris = datasets.load_iris()
print(dir(iris))
print(iris.target)
```

```python
# IRIS dataset separated as features and Targets

X = iris.data[:, :]
print(len(X))
y = iris.target
features = iris.feature_names
print(features)
print(X)
print(type(X))
print(y)
```

```python
# Finding the optimum number of clusters for k-means classification
# Using the elbow method

from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
#Plotting the results onto a line graph, allowing us to observe 'The elbow'
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```

```python
# Plot Between Sepal Length and Sepal Width

plt.scatter(X[:,0], X[:,1], c=y, cmap='gist_rainbow')
plt.xlabel('Sepa1 Length', fontsize=18)
plt.ylabel('Sepal Width', fontsize=18)
```

```python
# Plot Between Petal length and Petal Width

plt.scatter(X[:,2], X[:,3], c=y, cmap='gist_rainbow')
plt.xlabel('Petal Length', fontsize=18)
plt.ylabel('Petal Width', fontsize=18)
```

```python
# K Means Algorithm applied here, Task 1
# Clear from the elbow method, we need to have three clusters

kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)

y_kmeans = kmeans.fit_predict(X)
print(y_kmeans)
print(y)
```

```python
cluster_centers = kmeans.cluster_centers_
print(cluster_centers)
n_clusters=3
```

```python
# Separate the points based as per their cluster centers

def get_cluster_points(x1, x2, x3, x4):
    """
    Separate the cluster points specifically
    for the IRIS dataset
    """
    datapoints = []
    for i in range(0, len(x1)):
        datapoints.append([x1[i], x2[i], x3[i], x4[i]])
    return datapoints
    
x1_0 = X[y_kmeans == 0, 0]
x2_0 = X[y_kmeans == 0, 1]
x3_0 = X[y_kmeans == 0, 2]
x4_0 = X[y_kmeans == 0, 3]

x1_1 = X[y_kmeans == 1, 0]
x2_1 = X[y_kmeans == 1, 1]
x3_1 = X[y_kmeans == 1, 2]
x4_1 = X[y_kmeans == 1, 3]

x1_2 = X[y_kmeans == 2, 0]
x2_2 = X[y_kmeans == 2, 1]
x3_2 = X[y_kmeans == 2, 2]
x4_2 = X[y_kmeans == 2, 3]
```

```python
# Generic euclidean distance fucntion

def euclidean_distance(data_point, center):
    """
    Generic euclidean distance fucntion
    """
    distance = 0
    for i in range(0, len(data_point)):
        distance = distance + (data_point[i]-center[i])**2
    return math.sqrt(distance)

def get_center_by_average(data):
    """
    Generic average datapoint of all clusters
    This center will be almost same as the Kmeans calculated center
    Maybe a solution to the above doubt asked?
    """

    average_distance=[]
    total = len(data)
    data_length = data[0]
    
    for i in range(0, len(data_length)):
        dist=0
        aver=0
        for data_point in data:
            dist = dist+data_point[i]
        average_distance.append(dist/total)

    print(average_distance)
    return average_distance

def closest_center(average_center):
    """
    Return the closest center by checking
    the minimum distance
    """

    index_array = []
    for i in range(0, len(cluster_centers)):
        index_array.append([i, euclidean_distance(average_center, cluster_centers[i])])
    print(index_array)
    min1 = 1000000000
    for i in range(0, len(index_array)):
        min1 = min(min1, index_array[i][1])
    
    index=-1;
    for point in index_array:
        if point[1]==min1:
            index=point[0]
            break
    print(index)
    return cluster_centers[index]

def all_distance_from_clusters(center, data):
    """
    get euclidean distance from the center
    """

    dist_arr = []
    for point in data:
        dist_arr.append(euclidean_distance(point, center))
    return dist_arr
```

```python
# Distance between cluster centers and every data point in the dataset

print(kmeans.cluster_centers_)

get_center_by_average(get_cluster_points(x1_0, x2_0, x3_0, x4_0))
get_center_by_average(get_cluster_points(x1_1, x2_1, x3_1, x4_1))
get_center_by_average(get_cluster_points(x1_2, x2_2, x3_2, x4_2))

closest_center(get_center_by_average(get_cluster_points(x1_0, x2_0, x3_0, x4_0)))
```

```python
# Task 2, get all the distances from it's corresponding center

all_0 = all_distance_from_clusters(
    closest_center(get_center_by_average(get_cluster_points(x1_0, x2_0, x3_0, x4_0))),
    get_cluster_points(x1_0, x2_0, x3_0, x4_0),
)

all_1 = all_distance_from_clusters(
    closest_center(get_center_by_average(get_cluster_points(x1_1, x2_1, x3_1, x4_1))),
    get_cluster_points(x1_1, x2_1, x3_1, x4_1),
)

all_2 = all_distance_from_clusters(
    closest_center(get_center_by_average(get_cluster_points(x1_2, x2_2, x3_2, x4_2))),
    get_cluster_points(x1_2, x2_2, x3_2, x4_2),
)

min0 = np.min(all_0)
max0 = np.max(all_0)
print(max0)

min1 = np.min(all_1)
max1 = np.max(all_1)
print(max1)

min2 = np.min(all_2)
max2 = np.max(all_2)

print(min0, max0)
print(min1, max1)
print(min2, max2)

max_min_arr = []
max_min_arr.append([min0, max0])
max_min_arr.append([min1, max1])
max_min_arr.append([min2, max2])
# Now find maximum and minumum distance from each cluster
```

```python
# get variance
# as we have seen that the distance between Kmean center and the
# average center is negligible, so considering average center
# to be the cluster center

var0 = np.var(all_0)
print(var0)

var1 = np.var(all_1)
print(var1)

var2 = np.var(all_2)
print(var2)

var_arr = []
var_arr.append(var0)
var_arr.append(var1)
var_arr.append(var2)

# Calculate number of points within the 90 % of the cluster
# in all the clusters

def num_of_points_within(max1, data):
    """
    number of points within the 90 % of the cluster
    in all the clusters
    """
    tot = 0
    for dist in data:
        if dist<(0.9*max1):
            tot=tot+1
    print(tot)
    return tot

num_0 = num_within_0 = num_of_points_within(
    max0,
    all_0
)
num_1 = num_within_1 = num_of_points_within(
    max1,
    all_1
)
num_2 = num_within_2 = num_of_points_within(
    max2,
    all_2
)

num_within = []
num_within.append([num_0, len(all_0)])
num_within.append([num_1, len(all_1)])
num_within.append([num_2, len(all_2)])
```

```python
lab=kmeans.labels_
from sklearn.metrics import silhouette_score

sc=silhouette_score(X, kmeans.labels_)
print(sc)
```
